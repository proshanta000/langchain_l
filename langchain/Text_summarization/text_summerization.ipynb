```python
from langchain_classic.chains.llm import LLMChain
from langchain_core.prompts import PromptTemplate

generictemplate="""
Write a summery of the following speech.
speech:{speech}
Translate the precise summery to {language}
"""

prompt=PromptTemplate(
    input_variables=['speech', 'language'],
    template=generictemplate
)
prompt
```

```python
complete_prompt =prompt.format(speech=speech, language="Bangla")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
translation_chain = prompt | llm

# Run the chain using .invoke()
# Note: Since the model output is an AIMessage, we use .content at the end.
summary_and_translation = translation_chain.invoke({
    "speech": speech,
    "language": "Bangla"
}).content

print(summary_and_translation)
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("small.pdf")
docs = loader.load_and_split()
docs
```

```python
template="""
Write a concise and short summery of the following report,
report:{text}
"""

prompt = PromptTemplate(
    input_variables=["text"],
    template=template
    )
```

```python
from langchain_classic.chains.summarize import load_summarize_chain
```

```python

chain =load_summarize_chain(
    llm, chain_type="stuff", prompt=prompt,verbose=True 
)
```

```python
output_summery=chain.run(docs)
output_summery
```

```python
## Map reduce for large documents
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_spilitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100)
final_documents=text_spilitter.split_documents(docs)
final_documents
```

```python
len(final_documents)
```

```python
# 1. Define the MAP prompt (used for summarizing each chunk)
map_prompt_template_str = """
Please summarize the following report section:
Report section: {text}
CONCISE SUMMARY:
"""
MAP_PROMPT_TEMPLATE = PromptTemplate(
    input_variables=['text'],
    template=map_prompt_template_str
)
```

```python
# 2. Define the FINAL/COMBINE prompt (used for merging all chunk summaries)
final_prompt_str = """
Provide the final summary of the entire report with these important points.
Add a Motivational Title, Start the precise summary with an introduction and provide the summary in number
points for the report.
Report sections summaries: {text}
"""
FINAL_PROMPT_TEMPLATE = PromptTemplate(
    input_variables=['text'],
    template=final_prompt_str # Ensure you use 'template' here
)
```

```python
# 3. Load the Summarize Chain
summary_chain = load_summarize_chain(
    llm=llm,
    chain_type='map_reduce',
    map_prompt=MAP_PROMPT_TEMPLATE,          # Pass the MAP PromptTemplate object
    combine_prompt=FINAL_PROMPT_TEMPLATE,    # Pass the COMBINE PromptTemplate object
    verbose=True
)
```

```python
# 4. Run the Chain
summary_and_translation = summary_chain.invoke({"input_documents": docs})
final_output = summary_and_translation['output_text']

print("\n--- Summary (using map_reduce) ---")
print(final_output)
```
